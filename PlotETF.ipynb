```python
#import Data_generator_multiple as dg
import numpy as np
import math, random
import matplotlib.pyplot as plt
```

```python
# Data for plotting
x = np.array([0.0005,0.0025,0.005,0.01,0.025])
ESR_RNN = np.array([0.0725,0.0725,0.0724,0.0724,0.0723])
ESR_NTR = np.array([0.0724,0.0723,0.0723,0.0722,0.0721])
ESR_ETF = np.array([0.0721,0.0704,0.0683,0.0642,0.0512])
lower = np.array([0.07089633,0.07090398,0.07086611,0.07079167,0.07090988])

upper = np.array([0.07411097,0.07410813,0.07404821,0.07400747,0.07414206])

fig, ax = plt.subplots()
ax.set_xscale('log')
ax.set_yscale('linear')
RNN, = ax.plot(x, ESR_RNN, label='ESR_RNN',color='blue',  
     linewidth=1, markersize=12)
NTR, = ax.plot(x, ESR_NTR, label='ESR_NTR',color='red',  
     linewidth=1, markersize=12)
ETF, = ax.plot(x, ESR_ETF, label='ESR_ETF',color='green',  
     linewidth=1, markersize=12)
CI = ax.fill_between(x, lower, upper, alpha=0.2)
#ax.plot(x, y2, label='Simulated',color='blue',linestyle='dashed', 
     #linewidth=1, markersize=12)
#ax.set_ylim([-4.62, -4.59])
#ax.plot(t.cpu().detach(), s3.cpu().view(-1).detach())
#Cmean, = ax.plot(x, (lower+upper)/2, 'k--',color='blue')
ax.legend([RNN, (CI), NTR, ETF], ["ESR_RNN", "CI", "NTR", "ETF"], loc=1)

ax.set(xlabel=r'$\epsilon$', 
       title='Equivalent Safe Rate')
ax.set_ylabel(r'ESR',labelpad = 0)
ax.grid()

fig.savefig("ETF.png",dpi=500,pad_inches = 0.1,bbox_inches = 'tight')

plt.show()
```
