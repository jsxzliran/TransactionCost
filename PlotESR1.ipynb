```python
#import Data_generator_multiple as dg
import numpy as np
import torch
import torch.nn as nn
from torch.nn import functional as F
from torch.autograd import Variable
from torch import optim
import math, random
import matplotlib.pyplot as plt
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
```

```python
# Data for plotting
x = np.array([0.001, 0.005, 0.01, 0.05])
theo = np.array([0.026497745209933064, 0.026172737330812847, 0.025882602719334357, 0.02437404977677379])

lower = np.array([0.025110521737108833, 0.024592297072273443, 0.024140520196777072, 0.02206897283466805])

upper = np.array([0.02704056919658025, 0.02649295702933563, 0.026014330028561582, 0.023820647165796644])

fig, ax = plt.subplots()
ax.set_xscale('log')
ax.set_yscale('log')
theoretical, = ax.plot(x, theo, label='Theoretical',color='red',  
     linewidth=1, markersize=12)
CI = ax.fill_between(x, lower, upper, alpha=0.2)
#ax.plot(x, y2, label='Simulated',color='blue',linestyle='dashed', 
     #linewidth=1, markersize=12)
#ax.set_ylim([-4.62, -4.59])
#ax.plot(t.cpu().detach(), s3.cpu().view(-1).detach())
Cmean, = ax.plot(x, (lower+upper)/2, 'k--',color='blue')
ax.legend([theoretical, (CI,Cmean)], ["Theoretical", "CI+RNN"], loc=1)

ax.set(xlabel=r'$\epsilon$', 
       title='Equivalent Safe Rate')
ax.set_ylabel(r'$ESR$',labelpad = 0)
ax.grid()
from google.colab import files
fig.savefig("ESR_1.png",dpi=500,pad_inches = 0.1,bbox_inches = 'tight')
files.download("ESR_1.png")
plt.show()
```
